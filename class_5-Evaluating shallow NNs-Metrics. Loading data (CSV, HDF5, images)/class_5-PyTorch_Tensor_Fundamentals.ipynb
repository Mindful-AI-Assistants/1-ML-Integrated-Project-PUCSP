### Conversão entre PyTorch e NumPy
Podemos converter facilmente um tensor PyTorch em um array NumPy usando a função .numpy(). Da mesma forma, podemos converter um array NumPy em um tensor PyTorch usando a função torch.from_numpy().

```python
# Numpy
# Converting a Torch Tensor to a NumPy array and vice versa is very easy
a = torch.ones(5)
print(a)

# torch to numpy with .numpy()
b = a.numpy()
print(b)
print(type(b))

# Carful: If the Tensor is on the CPU (not the GPU),
# both objects will share the same memory location, so changing one
# will also change the other
a.add_(1)
print(a)
print(b)
```

```python
# numpy to torch with .from_numpy(x)
import numpy as np
a = np.ones(5)
b = torch.from_numpy(a)
print(a)
print(b)

# again be careful when modifying
a += 1
print(a)
print(b)
```

## Autograd

Esse código mostra exemplos de como usar o PyTorch com a funcionalidade de diferenciação automática fornecida pelo pacote autograd.

O código começa criando um tensor x de forma aleatória e com um sinalizador `requires_grad=True` que permite rastrear todas as operações executadas sobre esse tensor. Então, é criado outro tensor y que é resultado de uma operação com o tensor x. Como o tensor y foi criado como resultado de uma operação, ele tem um atributo grad_fn que se refere a uma função que criou o tensor.

São realizadas mais operações com o tensor y, incluindo o cálculo dos gradientes com a função `backward()` que computa automaticamente as derivadas parciais em relação ao tensor de entrada.

O código também mostra como lidar com tensores que não rastreiam o histórico de operações e como lidar com tensores que possuem saída não escalar. Além disso, há um exemplo de como esvaziar os gradientes antes de uma nova etapa de otimização.

Por fim, o código menciona a existência de uma classe Optimizer em PyTorch que tem um método `zero_grad()` que pode ser usado para esvaziar os gradientes antes de uma nova etapa de otimização.


O pacote autograd fornece diferenciação automática para todas as operações em tensores

```python
import torch

# O requires_grad = True rastreia todas as operações no tensor.
x = torch.randn(3, requires_grad=True)
y = x + 2

# y foi criado como resultado de uma operação, então ele tem um atributo grad_fn.
# grad_fn referencia uma Função que criou o Tensor
print(x) # criado pelo usuário -> grad_fn é None
print(y)
print(y.grad_fn)

# Faça mais operações em y
z = y * y * 3
print(z)
z = z.mean()
print(z)
```

Vamos calcular os gradientes com backpropagation
Quando terminarmos nossa computação, podemos chamar .backward() e ter todos os gradientes calculados automaticamente.
O gradiente para este tensor será acumulado no atributo .grad.
É a derivada parcial da função em relação ao tensor

```python
z.backward()
print(x.grad) # dz/dx

# Em geral, torch.autograd é um mecanismo para computar o produto vetor-Jacobiano
# Ele computa derivadas parciais enquanto aplica a regra da cadeia
```

Modelo com saída não escalar:
Se um Tensor não é escalar (mais de um elemento), precisamos especificar argumentos para backward ()
especificar um argumento de gradiente que é um tensor de forma correspondente.
necessário para o produto vetor-Jacobiano

```python
x = torch.randn(3, requires_grad=True)

y = x * 2
for _ in range(10):
    y = y * 2

print(y)
print(y.shape)

v = torch.tensor([0.1, 1.0, 0.0001], dtype=torch.float32)
y.backward(v)
print(x.grad)
```

Pare um tensor de rastrear a história:
Por exemplo, durante nosso loop de treinamento, quando queremos atualizar nossos pesos
então esta operação de atualização não deve fazer parte do cálculo do gradiente
- x.requires_grad_(False)
- x.detach()
- wrap in 'with torch.no_grad():'

```python
# .requires_grad_(...) muda uma flag existente inplace.
a = torch.randn(2, 2)
print(a.requires_grad)
b = ((a * 3) / (a - 1))
print(b.grad_fn)
a.requires_grad_(True)
print(a.requires_grad)
b = (a * a).sum()
print(b.grad_fn)
```

```python
# .detach(): obtenha um novo Tensor com o mesmo conteúdo, mas sem cálculo de gradiente:
a = torch.randn(2, 2, requires_grad=True)
print(a.requires_grad)
b = a.detach()
print(b.requires_grad)
```

```python
# wrap in 'with torch.no_grad():'
a = torch.randn(2, 2, requires_grad=True)
print(a.requires_grad)
with torch.no_grad():
    print((x ** 2).requires_grad)
```

 backward() acumula o gradiente para este tensor no atributo .grad.
* Precisamos ter cuidado durante a otimização!

Use .zero_() para esvaziar os gradientes antes de uma nova etapa de otimização!

```python
weights = torch.ones(4, requires_grad=True)

for epoch in range(3):
    # apenas um exemplo simples
    model_output = (weights*3).sum()
    model_output.backward()

    print(weights.grad)
    # otimizar o modelo, ou seja, ajustar os pesos ...
    with torch.no_grad():
        weights -= 0.1 * weights.grad

    # isso é importante! Isso afeta os pesos e a saída final
    weights.grad.zero_()

print(weights)
print(model_output)

# Otimizador tem um método zero_grad ()
# optimizer = torch.optim.SGD([weights], lr=0.1)
# Durante o treinamento:
# optimizer.step()
# optimizer.zero_grad()
```

## Backpropagation

O código é um exemplo simples de otimização de parâmetros usando PyTorch. Ele segue os seguintes passos:

1. Inicializa os tensores x, y e w.
2. Define o tensor w como o parâmetro que será otimizado, usando o argumento requires_grad=True.
3. Realiza a forward pass para calcular a perda (loss) com base na previsão (y_predicted) e na saída esperada (y).
4. Realiza a backward pass para calcular os gradientes da perda em relação aos parâmetros.
5. Atualiza os pesos (w) usando um otimizador com base nos gradientes calculados.
6. Zera os gradientes antes de realizar a próxima iteração do processo de otimização.

O código também imprime o valor da perda e do gradiente para o parâmetro w a cada iteração.

```python
import torch

x = torch.tensor(1.0)
y = torch.tensor(2.0)

# Este é o parâmetro que queremos otimizar -> requires_grad=True
w = torch.tensor(1.0, requires_grad=True)

# forward pass para calcular a perda
y_predicted = w * x
loss = (y_predicted - y)**2
print(loss)
```

$$\frac{\partial \mathcal{Loss}}{\partial w} = 2 x (wx-y)$$ 

```python
# backward pass para calcular o gradiente dLoss/dw
loss.backward()
print(w.grad)
```

```python
# atualiza os pesos
# próxima forward e backward pass...

# continua otimizando:
# atualiza os pesos, esta operação não deve fazer parte do grafo computacional
with torch.no_grad():
    w -= 0.01 * w.grad
# não se esqueça de zerar os gradientes
w.grad.zero_()
```

## Gradiente descendente

### Computar cada passo manualmente

O primeiro bloco de código implementa uma regressão linear em numpy, onde a relação entre as entradas X e saídas Y é f = 2x. A implementação envolve o cálculo da saída do modelo (forward), da função de perda (loss) e do gradiente da função de perda em relação aos pesos (gradient). O treinamento é realizado iterativamente atualizando os pesos com base nos gradientes calculados.

```python
import numpy as np

# Regressão linear
# f = w * x

# aqui: f = 2 * x
X = np.array([1, 2, 3, 4], dtype=np.float32)
Y = np.array([2, 4, 6, 8], dtype=np.float32)

w = 0.0

# saída do modelo
def forward(x):
    return w * x

# loss = MSE
def loss(y, y_pred):
    return ((y_pred - y)**2).mean()

# J = MSE = 1/N * (w*x - y)**2
# dJ/dw = 1/N * 2x(w*x - y)
def gradient(x, y, y_pred):
    return np.dot(2*x, y_pred - y).mean()

print(f'Predição antes do treinamento: f(5) = {forward(5):.3f}')

# Treinamento
learning_rate = 0.01
n_iters = 20

for epoch in range(n_iters):
    # predict = forward pass
    y_pred = forward(X)

    # loss
    l = loss(Y, y_pred)

    # calculate gradients
    dw = gradient(X, Y, y_pred)

    # update weights
    w -= learning_rate * dw

    if epoch % 2 == 0:
        print(f'época {epoch+1}: w = {w:.3f}, loss = {l:.8f}')

print(f'Predição após o treinamento: f(5) = {forward(5):.3f}')
```

### Autograd

O segundo bloco de código é uma implementação semelhante de regressão linear, mas usando o pacote PyTorch. Mais uma vez, a relação entre as entradas X e saídas Y é f = 2x. No entanto, o PyTorch simplifica muito o processo de treinamento em relação à implementação manual em numpy. O cálculo da saída do modelo, da função de perda e do gradiente são feitos automaticamente com o recurso de autograd. As atualizações de peso são feitas com o auxílio do otimizador de descida de gradiente em vez de atualizações manuais.


```python
import torch

# Regressão Linear
# f = w * x

# aqui: f = 2 * x
X = torch.tensor([1, 2, 3, 4], dtype=torch.float32)
Y = torch.tensor([2, 4, 6, 8], dtype=torch.float32)

w = torch.tensor(0.0, dtype=torch.float32, requires_grad=True)

# saída do modelo
def forward(x):
    return w * x

# loss = MSE
def loss(y, y_pred):
    return ((y_pred - y)**2).mean()

print(f'Previsão antes do treinamento: f(5) = {forward(5).item():.3f}')

# Treinamento
learning_rate = 0.01
n_iters = 100

for epoch in range(n_iters):
    # previsão = forward pass
    y_pred = forward(X)

    # loss
    l = loss(Y, y_pred)

    # calcular gradientes = backward pass
    l.backward()

    # atualizar pesos
    with torch.no_grad():
        w -= learning_rate * w.grad

    # zerar os gradientes após a atualização
    w.grad.zero_()

    if epoch % 10 == 0:
        print(f'epoch {epoch+1}: w = {w.item():.3f}, loss = {l.item():.8f}')

print(f'Previsão após o treinamento: f(5) = {forward(5).item():.3f}')
```
